```python
import os 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, Activation
import numpy as np
```

```python
x=tf.random.normal([200,300])
x.shape
```

```python
model=Sequential()

model.add(
    Conv2D(filters=100,
           kernel_size=(3,3),
           padding='SAME',
           strides=(2,2),
           input_shape=(200,300,3),
           activation='relu'
           )
)

model.add(
    Conv2D(filters=200,
           kernel_size=(3,3),
           padding='SAME',
           strides=(2,2),
           activation='relu'
           )
)

model.add(
    Conv2D(filters=400,
           kernel_size=(3,3),
           padding='SAME',
           strides=(2,2),
           activation='relu'
           )
)

model.summary()
```
